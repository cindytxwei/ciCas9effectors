```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib as mpl

plt.rcParams.update({'font.family':'arial'})

from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import LinearSegmentedColormap

%matplotlib inline
```

# HEK3 OT1

```python
df_A = pd.read_csv('Data/SPACE_ABE_HEK3_OT1.csv',index_col=0)

df_A = df_A*100

df_A.columns = ['C1','A2','C3','C4','C5','A6','G7','A8','C9','T10',
                'G11','A12','G13','C14','A15','C16','G17','T18','G19','C20']

df_A
```

```python
df_C = pd.read_csv('Data/SPACE_CBE_HEK3_OT1.csv',index_col=0)

df_C = df_C*100

df_C.columns = ['C1','A2','C3','C4','C5','A6','G7','A8','C9','T10',
                'G11','A12','G13','C14','A15','C16','G17','T18','G19','C20']

df_C
```

```python
fig, ax = plt.subplots(2,20,gridspec_kw={'height_ratios':[1,9]},figsize=(7.9,4))

fig.subplots_adjust(hspace=0)
fig.subplots_adjust(wspace=0)

ref = 'CACCCAGACTGAGCACGTGC'
zero = [0] * len(df_A)

ax_count = 20

cmap1 = LinearSegmentedColormap.from_list('mycmap', ['white', 'mediumvioletred'])
cmap1.set_bad(color='gainsboro')
cmap2 = LinearSegmentedColormap.from_list('mycmap', ['white', 'mediumblue'])
cmap2.set_bad(color='gainsboro')

for i in range(0,20):
    if ref[i] == 'A':
        fig.axes[i].yaxis.set_visible(False)
        fig.axes[i].xaxis.set_visible(False)
        fig.axes[i].scatter(1,0,s=450,c='mediumseagreen',marker='s')
        fig.axes[i].text(1,0, 'A', ha='center',va='center',color='white',size=20)
        fig.axes[i].spines['top'].set_visible(False)
        fig.axes[i].spines['bottom'].set_visible(False)
        fig.axes[i].spines['left'].set_visible(False)
        fig.axes[i].spines['right'].set_visible(False)
        fig.axes[ax_count].yaxis.set_visible(False)
        fig.axes[ax_count].xaxis.set_visible(False)
        df_plt = np.expand_dims(df_A.iloc[:,i], axis=0)
        df_plt2 = np.ma.masked_where(df_plt < 0.1, df_plt)
        g1 = fig.axes[ax_count].imshow(df_plt2.T,cmap=cmap1,aspect="auto",interpolation='none',vmin=0,vmax=2)
        fig.axes[ax_count].spines['top'].set_visible(False)
        fig.axes[ax_count].spines['bottom'].set_visible(False)
        fig.axes[ax_count].spines['left'].set_visible(False)
        fig.axes[ax_count].spines['right'].set_visible(False)
        fig.axes[ax_count].hlines([-0.5,0.5,1.5,3.5,4.5,6.5,7.5,8.5], xmin=-0.5, xmax=0.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].vlines([-0.5,0.5], ymin=-0.5, ymax=8.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].hlines([2.5,5.5], xmin=-0.5, xmax=0.5, color="black")
    elif ref[i] == 'C':
        fig.axes[i].yaxis.set_visible(False)
        fig.axes[i].xaxis.set_visible(False)
        fig.axes[i].scatter(1,0,s=450,c='royalblue',marker='s')
        fig.axes[i].text(1,0, 'C', ha='center',va='center',color='white',size=20)
        fig.axes[i].spines['top'].set_visible(False)
        fig.axes[i].spines['bottom'].set_visible(False)
        fig.axes[i].spines['left'].set_visible(False)
        fig.axes[i].spines['right'].set_visible(False)
        fig.axes[ax_count].yaxis.set_visible(False)
        fig.axes[ax_count].xaxis.set_visible(False)
        df_plt = np.expand_dims(df_C.iloc[:,i], axis=0)
        df_plt2 = np.ma.masked_where(df_plt < 0.1, df_plt)
        g2 = fig.axes[ax_count].imshow(df_plt2.T,cmap=cmap2,aspect="auto",interpolation='none',vmin=0,vmax=5)
        fig.axes[ax_count].spines['top'].set_visible(False)
        fig.axes[ax_count].spines['bottom'].set_visible(False)
        fig.axes[ax_count].spines['left'].set_visible(False)
        fig.axes[ax_count].spines['right'].set_visible(False)
        fig.axes[ax_count].hlines([-0.5,0.5,1.5,3.5,4.5,6.5,7.5,8.5], xmin=-0.5, xmax=0.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].vlines([-0.5,0.5], ymin=-0.5, ymax=8.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].hlines([2.5,5.5], xmin=-0.5, xmax=0.5, color="black")
    elif ref[i] == 'G':
        fig.axes[i].yaxis.set_visible(False)
        fig.axes[i].xaxis.set_visible(False)
        fig.axes[i].scatter(1,0,s=450,c='gold',marker='s')
        fig.axes[i].text(1,0, 'G', ha='center',va='center',color='white',size=20)
        fig.axes[i].spines['top'].set_visible(False)
        fig.axes[i].spines['bottom'].set_visible(False)
        fig.axes[i].spines['left'].set_visible(False)
        fig.axes[i].spines['right'].set_visible(False)
        fig.axes[ax_count].yaxis.set_visible(False)
        fig.axes[ax_count].xaxis.set_visible(False)
        df_A.iloc[:,i] = zero
        df_plt = np.expand_dims(df_A.iloc[:,i], axis=0)
        df_plt2 = np.ma.masked_where(df_plt < 0.1, df_plt)
        fig.axes[ax_count].imshow(df_plt2.T,cmap=cmap1,aspect="auto",interpolation='none',vmin=0,vmax=65)
        fig.axes[ax_count].spines['top'].set_visible(False)
        fig.axes[ax_count].spines['bottom'].set_visible(False)
        fig.axes[ax_count].spines['left'].set_visible(False)
        fig.axes[ax_count].spines['right'].set_visible(False)
        fig.axes[ax_count].hlines([-0.5,0.5,1.5,3.5,4.5,6.5,7.5,8.5], xmin=-0.5, xmax=0.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].vlines([-0.5,0.5], ymin=-0.5, ymax=8.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].hlines([2.5,5.5], xmin=-0.5, xmax=0.5, color="black")
    elif ref[i] == 'T':
        fig.axes[i].yaxis.set_visible(False)
        fig.axes[i].xaxis.set_visible(False)
        fig.axes[i].scatter(1,0,s=450,c='tomato',marker='s')
        fig.axes[i].text(1,0, 'T', ha='center',va='center',color='white',size=20)
        fig.axes[i].spines['top'].set_visible(False)
        fig.axes[i].spines['bottom'].set_visible(False)
        fig.axes[i].spines['left'].set_visible(False)
        fig.axes[i].spines['right'].set_visible(False)
        fig.axes[ax_count].yaxis.set_visible(False)
        fig.axes[ax_count].xaxis.set_visible(False)
        df_C.iloc[:,i] = zero
        df_plt = np.expand_dims(df_C.iloc[:,i], axis=0)
        df_plt2 = np.ma.masked_where(df_plt < 0.1, df_plt)
        fig.axes[ax_count].imshow(df_plt2.T,cmap=cmap1,aspect="auto",interpolation='none',vmin=0,vmax=65)
        fig.axes[ax_count].spines['top'].set_visible(False)
        fig.axes[ax_count].spines['bottom'].set_visible(False)
        fig.axes[ax_count].spines['left'].set_visible(False)
        fig.axes[ax_count].spines['right'].set_visible(False)
        fig.axes[ax_count].hlines([-0.5,0.5,1.5,3.5,4.5,6.5,7.5,8.5], xmin=-0.5, xmax=0.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].vlines([-0.5,0.5], ymin=-0.5, ymax=8.5, color="grey", linewidth=0.75)
        fig.axes[ax_count].hlines([2.5,5.5], xmin=-0.5, xmax=0.5, color="black")
    ax_count += 1
    
cbar_ax = fig.add_axes([0.92, 0.12, 0.02, 0.68])
cbar1 = fig.colorbar(g2, cax=cbar_ax)
cbar1.set_label('C > T editing frequency (%)',fontsize=16)
cbar1.ax.tick_params(labelsize=16)

cbar_ax = fig.add_axes([1.05, 0.12, 0.02, 0.68])
cbar2 = fig.colorbar(g1, cax=cbar_ax)
cbar2.set_label('A > G editing frequency (%)',fontsize=16)
cbar2.ax.tick_params(labelsize=16)

plt.savefig('ExportedPlots/SPACE_HEK3-OT1_heatmap_20211116.svg',dpi=300,bbox_inches='tight')
```

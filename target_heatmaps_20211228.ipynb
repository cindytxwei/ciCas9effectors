```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.rcParams.update({'font.family':'arial'})

from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import LinearSegmentedColormap

%matplotlib inline
```

```python
df = pd.read_csv('Data/BE4max_HEK3.csv',index_col=0)

df.columns = ['G1','G2','C3','C4','C5','A6','G7','A8','C9','T10',
              'G11','A12','G13','C14','A15','C16','G17','T18','G19','A20']

df = df*100

df
```

```python
df2 = df.div(df.C5, axis=0) * 100

df2
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,gridspec_kw={'height_ratios':[1,9]},figsize=(8,4))

#plt.setp(ax1.get_xticklabels(), visible=False)
fig.subplots_adjust(hspace=0)

ax1.set_ylim(-0.1,0.1)
ax1.set_xlim(-0.5,19.5)
ax2.set_xlim(-0.5,19.5)

ax1.axes.yaxis.set_visible(False)
ax1.axes.xaxis.set_visible(False)

ax1.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['right'].set_visible(False)

plt.setp(ax2.spines.values(), color='grey')

ref = 'GGCCCAGACTGAGCACGTGA'
zero = [0] * len(df)

for x,nucleotide in enumerate(ref):
    if nucleotide == 'A':
        ax1.scatter(x,0,s=450,c='mediumseagreen',marker='s')
        ax1.text(x,0, 'A', ha='center',va='center',color='white',size=20)
    elif nucleotide == 'C':
        ax1.scatter(x,0,s=450,c='royalblue',marker='s')
        ax1.text(x,0, 'C', ha='center',va='center',color='white',size=20)
    elif nucleotide == 'G':
        ax1.scatter(x,0,s=450,c='gold',marker='s')
        ax1.text(x,0, 'G', ha='center',va='center',color='white',size=20)
    elif nucleotide == 'T':
        ax1.scatter(x,0,s=450,c='tomato',marker='s')
        ax1.text(x,0, 'T', ha='center',va='center',color='white',size=20)
        n = df.columns[x]
        df.drop(n,axis=1,inplace=False)
        df[n] = zero

cmap = LinearSegmentedColormap.from_list('mycmap', ['white', '#990000'])
masked_array = np.ma.masked_where(df < 0.7, df2)
cmap.set_bad(color='gainsboro')

g2 = ax2.imshow(masked_array,cmap=cmap,aspect="auto",interpolation='none',vmin=0,vmax=100)

cbar_ax = fig.add_axes([0.92, 0.12, 0.02, 0.68])
cbar = fig.colorbar(g2, cax=cbar_ax)
cbar.set_label('% of maximum\nC > T editing',fontsize=16)
cbar.ax.tick_params(labelsize=16)

ax2.set_xticks(np.arange(0, 20, 1))
ax2.set_xticklabels(np.arange(1, 21, 1), fontsize=14)
ax2.set_xticks(np.arange(-.5, 20, 1), minor=True)
ax2.set_yticks(np.arange(-.5, 9, 1), minor=True)
ax2.grid(which='minor', color='grey', linestyle='-', linewidth=0.75)

ax2.hlines([2.5,5.5], xmin=-0.5, xmax=19.5, color="black")

ax2.xaxis.set_ticks_position('none')
ax2.yaxis.set_ticks_position('none')

ytick_labels = ['','','control','','','BE4max','','','ciBE4max']
ax2.set_yticklabels(labels = ytick_labels,
                    fontsize = 16)

ax1.set_title('HEK3', fontsize=20)

plt.savefig('ExportedPlots/BE4max_HEK3_fulltarget_heatmap_20211228.pdf',dpi=300,bbox_inches='tight')
```
